# -> Introduction to Repetition in Programming

## 1️⃣ What is Repetition?

#### Repetition in programming refers to executing a block of code multiple times until a certain condition is met.

#### In simple terms:

#### Repetition allows a computer to perform the same task again and again automatically.

#### Without repetition, programs would require writing the same code many times — which would be inefficient, error-prone, and unmanageable.

## 2️⃣ Why Repetition is Necessary

#### In real-world software systems, repetition is everywhere.

#### Consider these examples:

#### • Printing 100 student names

#### • Processing 1,000 transactions in a banking system

#### • Checking login attempts

#### • Sending emails to multiple users

#### • Reading thousands of data records

#### • Manually writing code for each repetition would be impossible.

#### Repetition makes programs:

#### • Efficient

#### • Scalable

#### • Clean

#### • Maintainable

## 3️⃣ Real-Life Analogy

#### Imagine telling someone:

#### “Write ‘I will practice Python’ 100 times.”

#### Manually writing it 100 times is repetitive and time-consuming.

#### Instead, you give an instruction:

#### “Repeat this sentence 100 times.”

#### That instruction is equivalent to a loop in programming.

###  Manual Repetition (Bad Practice)

```python
print("Welcome")
print("Welcome")
print("Welcome")
print("Welcome")
print("Welcome")
```

## ✅ Automated Repetition (Professional Way)

```python
for i in range(5):
    print("Welcome")
```

#### Each time the loop runs once, it is called an iteration.

## 6️⃣ How Repetition Works Internally (Logical Flow)

# -> What is a Loop?

## 1️⃣ Definition:

#### A loop is a programming construct that repeats a block of code multiple times until a specific condition is met.

#### Loops automate repetition so you don’t have to write the same code over and over.

## 2️⃣ Key Components of a Loop:

#### Every loop has three main components:

| Component          | Description                                                                 |
| ------------------ | --------------------------------------------------------------------------- |
| **Initialization** | A variable is set to a starting value before the loop starts.               |
| **Condition**      | A logical test that determines whether the loop continues or stops.         |
| **Update**         | Changes the loop variable after each iteration to eventually stop the loop. |


### Example (manual simulation):

```python
count = 1          # Initialization

while count <= 3:  # Condition
    print("Hello") # Loop body
    count += 1     # Update
```

## 3️⃣ Flow of Execution
### Step-by-step Logic

#### 1. Start the loop

#### 2. Check the loop condition

#### 3. If condition True → execute loop body

#### 4. Apply update (increment/decrement)

#### 5. Repeat step 2

#### 6. If condition False → exit loop

### 🔹 Flowchart Representation

        Start
          |
      Initialize
          |
      Check Condition
          |
    +-----+--------+
    |              |
    True          False
    |              |
    Execute Body  End Loop
    |
    Update
    |
    Back to Condition

#### This flowchart shows how loops continuously execute until the condition becomes False.

## 4️⃣ Loop Execution Example
### Example 1 – Counting Numbers 1 to 5

```python
num = 1           # Initialization

while num <= 5:   # Condition
    print(num)    # Loop body
    num += 1      # Update
```

### Example 2 – Loop with Conditional Inside

```python
num = 1

while num <= 5:
    if num % 2 == 0:
        print(num, "is Even")
    else:
        print(num, "is Odd")
    num += 1
```

#### This shows how loops and conditions can work together.

# -> Types of Loops in Python

#### Python has two main types of loops:

#### 1. for loop – used for count-controlled or iterable-based repetition

#### 2. while loop – used for condition-controlled repetition

#### Both loops can be nested, combined with conditional statements, and work with any iterable data structure.

## 1️⃣ for Loop:
### Definition:

#### A for loop in Python iterates over a sequence (like list, string, tuple, dictionary, or range) and executes the loop body once for each item.

### Syntax:

#### for variable in sequence:
####     # loop body

#### • variable → temporarily holds the current element

#### • sequence → the iterable object

#### • Indentation → defines loop body

### 🔹 Example 1 – Loop through Numbers

```python
for i in range(5):
    print("Iteration number:", i)
```

### 🔹 Example 2 – Loop through List

```python
fruits = ["Apple", "Banana", "Cherry"]

for fruit in fruits:
    print("I like", fruit)
```

### 🔹 Example 3 – Loop through Dictionary

```python
student_scores = {"Ali": 90, "Sara": 85, "Ahmed": 78}

for name, score in student_scores.items():
    print(name, "scored", score)
```

## 🔹 Flowchart for for Loop:

          Start
            |
    Initialize sequence
            |
    Take next element?
       +----+----+
       |         |
      Yes         No
       |          |
    Execute Body  End
       |
    Back to next element


## 2️⃣ while Loop:

### Definition:

#### A while loop executes a block of code as long as a condition is True.

### Syntax:

#### while condition:
####     # loop body

#### • Condition is checked before each iteration

#### • Loop stops when condition becomes False

### 🔹 Example 1 – Counting with While Loop

```python
count = 1

while count <= 5:
    print("Count is:", count)
    count += 1
```

### 🔹 Example 2 – Conditional Execution

```python
number = 10

while number > 0:
    if number % 2 == 0:
        print(number, "is Even")
    else:
        print(number, "is Odd")
    number -= 1
```

### 🔹 Flowchart for while Loop:

         Start
           |
      Check condition
        +----+----+
        |         |
      True       False
        |          |
    Execute Body   End
        |
    Update variable
        |
    Back to condition


### 3️⃣ Key Differences Between for and while

| Feature               | `for` Loop                    | `while` Loop                 |
| --------------------- | ----------------------------- | ---------------------------- |
| Iteration type        | Sequence / iterable           | Condition-based              |
| Use case              | Fixed number of iterations    | Unknown number of iterations |
| Initialization        | Optional inside loop          | Usually required before loop |
| Update                | Automatic (iterable moves)    | Manual update inside loop    |
| Risk of Infinite Loop | Low (range limits iterations) | High (wrong condition)       |


### 4️⃣ Practical Examples – When to Use Which Loop

#### • for loop: iterating over a list, dictionary, string, or range

#### • while loop: login retry attempts, countdowns, condition-based tasks

### Example – Login Attempt Using While Loop

```python
attempts = 0

while attempts < 3:
    password = input("Enter password: ")
    if password == "admin123":
        print("Login successful")
        break
    else:
        print("Wrong password")
        attempts += 1
```

# -> The range() Function

## 1️⃣ What is range()?

#### The range() function in Python is used to generate a sequence of numbers.

#### It is commonly used with for loops to control how many times a loop runs.

#### range() does not create a list.
#### It creates a special range object that generates numbers when needed.

## 2️⃣ Why range() is Important

#### Without range(), we cannot easily:

####  • Repeat something a specific number of times

####  • Generate number sequences

####  • Control loop counters

####  • Create numeric patterns

### Example without range() (inefficient):

```python
print(0)
print(1)
print(2)
print(3)
print(4)
```

### Now using range() (professional way):

```python
for i in range(5):
    print(i)
```

## 3️⃣ Syntax of range()

#### There are three forms of range():

### 🔹 1. range(stop)

#### range(stop)

#### • Starts from 0

#### • Ends at stop - 1

### Example:

```python
for i in range(5):
    print(i)
```

### 🔹 2. range(start, stop)

#### range(start, stop)


#### • Starts from start

#### • Stops before stop
### Example:

```python
for i in range(2, 7):
    print(i)
```

### 🔹 3. range(start, stop, step)

#### range(start, stop, step)


#### • Starts from start

#### • Stops before stop

#### • Increments by step

### Example:

```python
for i in range(1, 10, 2):
    print(i)
```

## 4️⃣ Important Rule: Stop Value is NOT Included

#### This is called exclusive stop.

### Example:

```python
for i in range(1, 5):
    print(i)
```

## 5️⃣ Negative Step (Reverse Looping)

#### We can count backwards using a negative step.

### Example:

```python
for i in range(5, 0, -1):
    print(i)
```

## 6️⃣ Flowchart of range() in a Loop:

          Start
            |
      Initialize start value
            |
      Check current < stop ?
          +----+----+
          |         |
          Yes       No
          |         |
    Execute body   End
          |
    Add step value
          |
    Back to condition


## 7️⃣ Converting range() to List

#### Since range() does not produce a list directly:

```python
numbers = range(5)
print(numbers)
```

#### To see actual numbers:

```python
numbers = list(range(5))
print(numbers)
```

## 8️⃣ Practical Example: Even Numbers Generator

```python
for i in range(2, 21, 2):
    print(i)
```

## 9️⃣ Using range() for Index Access

### Example:

```python
fruits = ["Apple", "Banana", "Cherry"]

for i in range(len(fruits)):
    print("Index:", i, "Value:", fruits[i])
```

# -> Loop Control Statements

#### Loop control statements allow us to change the normal flow of a loop.

#### Normally, a loop:

#### • Starts

#### • Checks condition

#### • Executes body

#### • Repeats until condition becomes False

####  But sometimes, we want to:

#### • Stop the loop immediately

#### • Skip a specific iteration

#### • Keep a placeholder for future logic

#### That’s where loop control statements are used.

#### Python provides three main loop control statements:

#### 1. break

#### 2. continue

#### 3. pass

## 1️⃣ The break Statement
### 🔹 Definition

#### The break statement is used to terminate the loop immediately, even if the loop condition is still True.

#### When break executes:

#### The loop stops instantly

#### Control moves to the first statement after the loop

### 🔹 Flowchart for break:

        Start Loop
            |
     Check Condition
            |
        Execute Body
            |
     Is break executed?
         +----+----+
         |         |
        Yes       No
         |         |
       Exit Loop   Continue


### 🔹 Example 1 – Basic break

```python
for i in range(1, 10):
    if i == 5:
        break
    print(i)
```

### 🔹 Example 2 – Real-Life Login System

```python
correct_password = "admin123"

for attempt in range(3):
    password = input("Enter password: ")
    
    if password == correct_password:
        print("Login successful")
        break
    else:
        print("Wrong password")
```

## 2️⃣ The continue Statement
### 🔹 Definition

#### The continue statement skips the current iteration and moves to the next iteration of the loop.

#### It does NOT stop the loop.
#### It only skips one cycle.

### 🔹 Flowchart for continue:

        Start Loop
            |
      Check Condition
            |
        Execute Body
            |
     Is continue executed?
     +---------+---------+
     |                   |
     Yes                 No
     |                   |
    Skip rest of body  Continue normally
         |
    Go to next iteration


### 🔹 Example 1 – Skip Even Numbers

```python
for i in range(1, 11):
    if i % 2 == 0:
        continue
    print(i)
```

### 🔹 Example 2 – Filtering Data

```python
numbers = [10, -5, 8, -2, 15]

for num in numbers:
    if num < 0:
        continue
    print("Valid number:", num)
```

## 3️⃣ The pass Statement
### 🔹 Definition

#### pass is a placeholder statement.
#### It does nothing.

#### It is used when:

#### • Syntax requires a statement

#### • But you do not want to write logic yet

### 🔹 Example 1 – Placeholder Loop

```python
for i in range(5):
    pass
```

### 🔹 Example 2 – Future Implementation

```python
for i in range(1, 6):
    if i == 3:
        pass
    else:
        print(i)
```

## 4️⃣ Comparison of Loop Control Statements:

| Statement | Stops Loop? | Skips Iteration? | Used As Placeholder? |
| --------- | ----------- | ---------------- | -------------------- |
| break     | Yes         | No               | No                   |
| continue  | No          | Yes              | No                   |
| pass      | No          | No               | Yes                  |


## 5️⃣ Combining Loop Control Statements:

### Example:

```python
for i in range(1, 11):
    if i == 8:
        break
    if i % 2 == 0:
        continue
    print(i)
```

# -> Nested Loops

## 1️⃣ What is a Nested Loop?

#### A nested loop is a loop inside another loop.

#### This means:

#### • The outer loop runs first.

#### • For each iteration of the outer loop,

#### • The inner loop runs completely.

#### Inner loop completes all its iterations for every single iteration of the outer loop.

## 2️⃣ Basic Structure of Nested Loops

```python
for outer in range(3):          # Outer loop
    for inner in range(2):      # Inner loop
        print("Outer:", outer, "Inner:", inner)
```

## 3️⃣ Execution Flow Explanation

#### Let’s trace the execution:

#### Outer loop runs 3 times → (0,1,2)

#### For each outer iteration:

#### • Inner loop runs 2 times → (0,1)

#### So total executions:

#### 3 × 2 = 6 times

## 4️⃣ Flowchart of Nested Loops:

            Start
              |
         Outer Loop Start
              |
      Check Outer Condition
              |
        Execute Inner Loop
              |
     Check Inner Condition
              |
      Execute Inner Body
              |
      Inner Loop Ends?
         +-----+-----+
         |           |
        No          Yes
         |           |
    Continue Inner    Back to Outer
                     |
             Outer Loop Ends?
                +----+----+
                |         |
               No        Yes
                |         |
         Repeat Process   End


## 5️⃣ Example 1 – Multiplication Table

```python
for i in range(1, 4):
    for j in range(1, 4):
        print(i, "x", j, "=", i*j)
    print("----")
```

## 6️⃣ Example 2 – Pattern Printing (Square Pattern)

```python
for i in range(4):
    for j in range(4):
        print("*", end=" ")
    print()
```

## 7️⃣ Example 3 – Triangle Pattern

```python
for i in range(1, 5):
    for j in range(i):
        print("*", end=" ")
    print()
```

## 8️⃣ Nested While Loop Example

```python
i = 1

while i <= 3:
    j = 1
    while j <= 2:
        print("i:", i, "j:", j)
        j += 1
    i += 1
```

# -> Infinite Loops

## 1️⃣ What is an Infinite Loop?

#### An infinite loop is a loop that never stops executing because its condition always remains True.

#### In simple words:

#### If the stopping condition never becomes False, the loop runs forever.

#### Infinite loops can:

#### • Freeze programs

#### • Crash systems

#### • Consume CPU resources

#### • Make software unresponsive

## 2️⃣ Basic Example of Infinite Loop

```python
while True:
    print("This will run forever")
```

### Why is this infinite?
#### 1. Condition is always True

#### 2. There is no break statement

#### 3. There is no stopping logic

## 3️⃣ Flowchart of Infinite Loop:

        Start
          |
     Check Condition
          |
        True
          |
     Execute Body
          |
     No Update
          |
     Back to Condition
          |
        True (Forever)


## 4️⃣ ️Real-World Example – ATM System

```python
while True:
    print("1. Withdraw")
    print("2. Deposit")
    print("3. Exit")
    
    choice = input("Enter choice: ")
    
    if choice == "3":
        print("Exiting system")
        break
```

## ️5️⃣ Debugging Infinite Loops

#### If program freezes:

#### • Press Ctrl + C in terminal

#### • Stop kernel in Jupyter Notebook

#### • Add debug prints inside loop

### Example debugging:

```python
count = 1

while count <= 5:
    print("Current count:", count)
    count += 1
```

# -> Advanced Loop Concepts

## 1️⃣ List Comprehension
### 🔹 What is it?

#### List comprehension is a shorter and more powerful way to create lists using loops.

### Traditional Way

```python
numbers = [1, 2, 3, 4, 5]
squares = []

for num in numbers:
    squares.append(num ** 2)

print(squares)
```

### List Comprehension Way

```python
numbers = [1, 2, 3, 4, 5]
squares = [num ** 2 for num in numbers]

print(squares)
```

### 🔹 With Condition

```python
numbers = [1, 2, 3, 4, 5, 6]

even_numbers = [num for num in numbers if num % 2 == 0]

print(even_numbers)
```

### 🔹 Nested List Comprehension

```python
matrix = [[1, 2], [3, 4], [5, 6]]

flattened = [num for row in matrix for num in row]

print(flattened)
```

## 2️⃣ Dictionary Comprehension

```python
numbers = [1, 2, 3, 4]

square_dict = {num: num ** 2 for num in numbers}

print(square_dict)
```

### 🔹 With Condition

```python
numbers = range(1, 6)

even_square = {num: num ** 2 for num in numbers if num % 2 == 0}

print(even_square)
```

## 3️⃣ Set Comprehension

```python
numbers = [1, 2, 2, 3, 3, 4]

unique_squares = {num ** 2 for num in numbers}

print(unique_squares)
```

### 🔹 Why Generators Are Important

## 7️⃣ Functional Loop Alternatives
### 🔹 map()

#### Applies function to each element.

```python
numbers = [1, 2, 3, 4]

def square(x):
    return x ** 2

result = list(map(square, numbers))

print(result)
```

### 🔹 filter()

#### Filters elements based on condition.

```python
numbers = [1, 2, 3, 4, 5]

def is_even(x):
    return x % 2 == 0

result = list(filter(is_even, numbers))

print(result)
```

# -> Real-World Case Studies Using Loops

## 1️⃣ Case Study: Student Result Analytics System
### 🎯 Problem

#### A school wants:

#### • Total marks

#### • Average marks

#### • Highest scorer

#### • Pass/Fail report

### 🔹 Theory

#### • Loops allow:

#### • Traversing student records

#### • Applying conditions

#### • Accumulating totals

#### • Finding maximum values

### 🔹 Implementation

```python
students = {
    "Ali": 85,
    "Sara": 42,
    "Ahmed": 91,
    "Zara": 67,
    "Usman": 39
}

total_marks = 0
highest_marks = 0
top_student = ""

for name, marks in students.items():
    total_marks += marks
    
    if marks > highest_marks:
        highest_marks = marks
        top_student = name

    if marks >= 50:
        print(name, "Passed")
    else:
        print(name, "Failed")

average = total_marks / len(students)

print("Average Marks:", average)
print("Top Student:", top_student)
```

## 2️⃣ Case Study: Login Authentication System
### 🎯 Problem

#### User gets 3 attempts to enter correct password.

### 🔹 Implementation

```python
correct_password = "python123"
attempts = 3

for attempt in range(attempts):
    entered = input("Enter password: ")

    if entered == correct_password:
        print("Access Granted")
        break
    else:
        print("Wrong Password")

else:
    print("Account Locked")
```
