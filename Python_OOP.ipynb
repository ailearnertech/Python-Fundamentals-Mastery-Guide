# -> Attributes in Classes 

## 1️⃣ What Are Attributes?

#### Attributes are:

#### Variables that belong to a class or an object.

#### They store data inside objects.

#### In real life:

### Car:

#### • color

#### • brand

#### • speed

### Student:

#### • name

#### • age

#### • marks

#### These are attributes.

#### In programming:

#### Attributes store object data.

## 2️⃣ Types of Attributes in Python

#### There are two main types:

### 1. Instance Variables

### 2. Class Variables

## 3️⃣ Instance Variables

#### Instance variables are:

#### Variables that belong to a specific object.

#### Each object has its own copy.

#### That means:

#### If we create 3 objects,
#### each object will have separate data.

### Example: Instance Variables

```python
# Creating a class
class Student:
    
    # Constructor to initialize instance variables
    def __init__(self, name, age):
        self.name = name  # Instance variable
        self.age = age    # Instance variable


# Creating first object
student1 = Student("Ali", 20)

# Creating second object
student2 = Student("Sara", 22)

# Printing values
print(student1.name)  # Ali
print(student1.age)   # 20

print(student2.name)  # Sara
print(student2.age)   # 22
```

## 4️⃣ Memory Visualization of Instance Variables

#### When we create:

#### student1 = Student("Ali", 20)

#### Memory looks like:

#### student1 → { name: "Ali", age: 20 }
#### student2 → { name: "Sara", age: 22 }

#### Each object has independent data.

## 5️⃣ Class Variables

#### Class variables are:

#### Variables that belong to the class itself.

#### They are shared by ALL objects.

#### Only one copy exists in memory.

### Example: Class Variable

```python
# Creating a class
class Student:
    
    school_name = "ABC School"  # Class variable (shared by all objects)
    
    def __init__(self, name):
        self.name = name  # Instance variable


# Creating objects
student1 = Student("Ali")
student2 = Student("Sara")

# Accessing class variable
print(student1.school_name)
print(student2.school_name)

# Accessing using class name
print(Student.school_name)
```

## 6️⃣ Difference Between Instance and Class Variables

| Feature        | Instance Variable        | Class Variable                    |
| -------------- | ------------------------ | --------------------------------- |
| Belongs To     | Object                   | Class                             |
| Memory         | Separate copy per object | Single shared copy                |
| Created Inside | Constructor (`__init__`) | Directly inside class             |
| Accessed By    | object.variable          | object.variable or Class.variable |


## 7️⃣ Modifying Class Variable

```python
class Student:
    
    school_name = "ABC School"
    
    def __init__(self, name):
        self.name = name


student1 = Student("Ali")
student2 = Student("Sara")

# Changing class variable
Student.school_name = "XYZ School"

print(student1.school_name)
print(student2.school_name)
```

## 8️⃣ What Happens If We Modify Using Object?

```python
class Student:
    
    school_name = "ABC School"
    
    def __init__(self, name):
        self.name = name


student1 = Student("Ali")
student2 = Student("Sara")

# Modifying using object
student1.school_name = "New School"

print(student1.school_name)
print(student2.school_name)
print(Student.school_name)
```

```python
print(student1.name)
print(Student.school_name)
```

### For Instance Variable:

        Object Created
              ↓
        __init__ runs

     self.variable created

### For Class Variable:

     Class Defined

## 1️⃣1️⃣ When to Use Instance Variable?

### Use when:

#### ✔ Data is different for each object
#### ✔ Example: name, age, marks

## 1️⃣2️⃣ When to Use Class Variable?

### Use when:

#### ✔ Data is common for all objects

## 1️⃣3️⃣ Complete Example:

```python
class Car:
    
    company = "Toyota"  # Class variable (shared)
    
    def __init__(self, model, color):
        self.model = model  # Instance variable
        self.color = color  # Instance variable


# Creating objects
car1 = Car("Corolla", "Black")
car2 = Car("Civic", "White")

# Printing instance variables
print(car1.model)
print(car2.model)

# Printing class variable
print(car1.company)
print(car2.company)
```

# -> Methods in Classes

## 1️⃣ What Are Methods?

#### A method is:

#### A function defined inside a class.

#### In simple words:

#### If attributes store data,
#### then methods perform actions.

## 2️⃣ Real-Life Understanding

#### Let’s take example of a Car.

#### Car has:

### Attributes (Data):

#### • color

#### • model

#### • speed

### Methods (Behavior):

#### • start()

### Without methods:

### Methods allow:

## 5️⃣ First Simple Method Example

```python
# Creating a class
class Student:
    
    def greet(self):  # Method inside class
        print("Hello Student!")  # Action performed


# Creating object
student1 = Student()

# Calling method using object
student1.greet()  # Method call
```

## 6️⃣ What is self in Methods?

#### self means:

#### The current object calling the method.

#### If student1 calls method,
#### then self = student1.

#### If student2 calls method,
#### then self = student2.

## 7️⃣ Method Using Instance Variables

#### Now let’s combine attributes + methods.

```python
class Student:
    
    def __init__(self, name):
        self.name = name  # Instance variable
    
    def display_name(self):
        print("Student Name is:", self.name)  # Accessing instance variable


# Creating objects
student1 = Student("Ali")
student2 = Student("Sara")

# Calling method
student1.display_name()
student2.display_name()
```

## 8️⃣ Methods with Parameters

#### Methods can also take extra parameters.

```python
class Calculator:
    
    def add(self, num1, num2):
        result = num1 + num2  # Adding numbers
        print("Result is:", result)


# Creating object
calc = Calculator()

# Calling method with arguments
calc.add(10, 5)
```

## 9️⃣ Method Modifying Object Data

#### Methods can change instance variables.

```python
class BankAccount:
    
    def __init__(self, balance):
        self.balance = balance  # Instance variable
    
    def deposit(self, amount):
        self.balance += amount  # Increasing balance
        print("New Balance:", self.balance)


# Creating object
account = BankAccount(1000)

# Calling method
account.deposit(500)
```

## 🔟 Calling One Method Inside Another

```python
class Demo:
    
    def first_method(self):
        print("First method called")
    
    def second_method(self):
        print("Second method called")
        self.first_method()  # Calling another method


# Creating object
obj = Demo()

# Calling second method
obj.second_method()
```

       Object Calls Method

               ↓

               ↓

               ↓

## 1️⃣2️⃣ Difference Between Function and Method

| Feature         | Function         | Method       |
| --------------- | ---------------- | ------------ |
| Defined         | Outside class    | Inside class |
| First Parameter | Normal parameter | `self`       |
| Called By       | Directly         | Using object |
| Belongs To      | Independent      | Class        |

## 1️⃣3️⃣ Complete Example 

```python
class Car:
    
    def __init__(self, model, speed):
        self.model = model  # Instance variable
        self.speed = speed  # Instance variable
    
    def display_info(self):
        print("Model:", self.model)
        print("Speed:", self.speed)
    
    def accelerate(self, increase):
        self.speed += increase  # Updating speed
        print("Speed increased to:", self.speed)


# Creating object
car1 = Car("Corolla", 100)

# Calling methods
car1.display_info()
car1.accelerate(20)
```

# -> The self Keyword in Python

## 1️⃣ What is self?

#### self is:

#### A reference to the current object.

#### In simple words:

#### When an object calls a method,
#### Python automatically sends that object to the method.

#### That object is received in the parameter called self.

## 2️⃣ Why Do We Need self?

#### Because:

#### Inside a class,
#### we can create multiple objects.

#### Each object has its own data.

#### So Python needs a way to know:

#### 👉 Which object is calling the method?

#### self solves this problem.

## 3️⃣ Simple Example Without Confusion

```python
class Student:
    
    def greet(self):  # self is required
        print("Hello Student!")


# Creating object
student1 = Student()

# Calling method
student1.greet()
```

## 4️⃣ Understanding self With Instance Variables

#### Now let’s use self to store and access data.

```python
class Student:
    
    def __init__(self, name):
        self.name = name  # Storing value inside object
    
    def display(self):
        print("Student Name:", self.name)  # Accessing object data


# Creating objects
student1 = Student("Ali")
student2 = Student("Sara")

student1.display()
student2.display()
```

## 5️⃣ What If We Do NOT Use self?

#### Let’s see wrong example:

```python
class Student:
    
    def __init__(self, name):
        name = name  # WRONG (no self)
    
    def display(self):
        print(name)  # ERROR


student1 = Student("Ali")
student1.display()
```

#### ❌ This will cause error.

### Why?

#### Because:

#### • name is local variable

#### • It is NOT stored in object

#### • It disappears after constructor finishes

#### That’s why we must use:

#### self.name
## 6️⃣ self Helps Differentiate Objects

### Example:

```python
class Car:
    
    def __init__(self, model):
        self.model = model
    
    def show(self):
        print("Model:", self.model)


car1 = Car("Corolla")
car2 = Car("Civic")

car1.show()
car2.show()
```

## 7️⃣ self in Multiple Methods

```python
class BankAccount:
    
    def __init__(self, balance):
        self.balance = balance
    
    def deposit(self, amount):
        self.balance += amount  # Using self to update data
    
    def show_balance(self):
        print("Balance:", self.balance)


account = BankAccount(1000)

account.deposit(500)
account.show_balance()
```

## 8️⃣ Flow of self (Very Important)

#### Flow when calling:

#### account.deposit(500)

### Flowchart:

       Object calls method
               ↓
    Python passes object automatically
               ↓
    self receives object reference
               ↓
    Method uses self to access object data

## 9️⃣ Important Rules About self

#### ✔ self is not a keyword
#### ✔ It is a convention (but must be used)
#### ✔ It must be first parameter of instance methods
#### ✔ It represents current object
#### ✔ Python passes it automatically

## 🔟 Can We Use Another Name Instead of self?

### Technically yes:

```python
class Demo:
    
    def greet(myobject):
        print("Hello")


obj = Demo()
obj.greet()
```

# -> Working with Multiple Objects & Object Behavior

## 1️⃣ Why Multiple Objects Are Important?

#### One of the biggest advantages of OOP is:

#### One class → Many objects

### In real world:

#### If we have a Student class,
#### we can create:

#### • 1 student

#### • 10 students

#### • 1000 students

#### Each student is a separate object.

## 2️⃣ Understanding Object Independence

### Every object:

#### ✔ Has its own memory
#### ✔ Has its own data
#### ✔ Can behave independently

#### Let’s prove this with code.

## 3️⃣ Basic Example of Multiple Objects

```python
# Creating a class
class Student:
    
    def __init__(self, name, marks):
        self.name = name      # Instance variable
        self.marks = marks    # Instance variable
    
    def display(self):
        print("Name:", self.name)
        print("Marks:", self.marks)
        print("-------------------")


# Creating multiple objects
student1 = Student("Ali", 85)
student2 = Student("Sara", 92)
student3 = Student("Ahmed", 78)

# Calling method for each object
student1.display()
student2.display()
student3.display()
```

## 4️⃣ Memory Visualization of Multiple Objects

#### When we create:

#### student1 = Student("Ali", 85)
#### student2 = Student("Sara", 92)

#### Memory looks like:

#### student1 → { name: "Ali", marks: 85 }
#### student2 → { name: "Sara", marks: 92 }

#### Separate memory blocks.

## 5️⃣ Object Behavior (How Objects Act)

#### Objects behave according to:

#### 1. Their data

#### 2. The methods defined in class

### Example:

```python
class BankAccount:
    
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance
    
    def deposit(self, amount):
        self.balance += amount
        print(self.name, "new balance:", self.balance)
    
    def withdraw(self, amount):
        self.balance -= amount
        print(self.name, "new balance:", self.balance)


# Creating multiple accounts
account1 = BankAccount("Ali", 1000)
account2 = BankAccount("Sara", 2000)

# Each object behaves independently
account1.deposit(500)
account2.withdraw(300)
```

## 6️⃣ Using Objects in a List

#### In real programs, we often store multiple objects in a list.

```python
class Student:
    
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks
    
    def display(self):
        print("Name:", self.name, "| Marks:", self.marks)


# Creating objects
student1 = Student("Ali", 85)
student2 = Student("Sara", 92)
student3 = Student("Ahmed", 78)

# Storing objects in a list
students = [student1, student2, student3]

# Looping through objects
for student in students:
    student.display()
```

## 7️⃣ Updating One Object Does NOT Affect Others

#### Let’s prove independence clearly.

```python
class Car:
    
    def __init__(self, model, speed):
        self.model = model
        self.speed = speed
    
    def accelerate(self, value):
        self.speed += value
    
    def show(self):
        print(self.model, "Speed:", self.speed)


# Creating two cars
car1 = Car("Corolla", 100)
car2 = Car("Civic", 120)

# Changing speed of only car1
car1.accelerate(20)

car1.show()
car2.show()
```

## Case Study 1: School Management System
### 🎯 Problem

#### A school wants to store:

#### • Student name

#### • Roll number

#### • Marks

#### And:

#### Display student details

#### Update marks

###  OOP Thinking

#### Real-world entity → Student

#### So we create:

#### Class → Student
#### Objects → Individual students

### Implementation

```python
# School Management System

class Student:
    
    # Constructor to initialize student data
    def __init__(self, name, roll_no, marks):
        self.name = name          # Instance variable
        self.roll_no = roll_no    # Instance variable
        self.marks = marks        # Instance variable
    
    # Method to display student details
    def display_details(self):
        print("Name:", self.name)
        print("Roll No:", self.roll_no)
        print("Marks:", self.marks)
        print("----------------------")
    
    # Method to update marks
    def update_marks(self, new_marks):
        self.marks = new_marks
        print("Marks updated successfully!")


# Creating student objects
student1 = Student("Ali", 101, 85)
student2 = Student("Sara", 102, 92)

# Display details
student1.display_details()
student2.display_details()

# Updating marks of student1
student1.update_marks(90)

# Display updated data
student1.display_details()
```

## Case Study 2: Bank Account System
### 🎯 Problem

#### Bank needs to:

#### • Store account holder name

#### • Store balance

#### • Allow deposit

#### • Allow withdrawal

#### • Show balance

###  OOP Thinking

#### Real-world entity → BankAccount

#### Each account is an object.

###  Implementation

```python
# Bank Account System

class BankAccount:
    
    # Constructor
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance
    
    # Deposit method
    def deposit(self, amount):
        self.balance += amount
        print(self.name, "deposited", amount)
    
    # Withdraw method
    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print(self.name, "withdrew", amount)
        else:
            print("Insufficient balance!")
    
    # Display balance
    def show_balance(self):
        print("Account Holder:", self.name)
        print("Current Balance:", self.balance)
        print("------------------------")


# Creating accounts
account1 = BankAccount("Ali", 5000)
account2 = BankAccount("Sara", 8000)

# Performing operations
account1.deposit(2000)
account1.withdraw(1000)
account1.show_balance()

account2.withdraw(9000)
account2.show_balance()
```

## Case Study 3: Car Rental System
### 🎯 Problem

#### Car rental company needs to:

#### • Store car model

#### • Store rent per day

#### • Calculate total rent

#### • Display car info

### OOP Thinking

#### Real-world entity → Car

#### Each car is an object.

###  Implementation

```python
# Car Rental System

class Car:
    
    # Constructor
    def __init__(self, model, rent_per_day):
        self.model = model
        self.rent_per_day = rent_per_day
    
    # Method to calculate total rent
    def calculate_rent(self, days):
        total = self.rent_per_day * days
        print("Total Rent for", days, "days is:", total)
    
    # Method to display car details
    def display_info(self):
        print("Car Model:", self.model)
        print("Rent per Day:", self.rent_per_day)
        print("----------------------")


# Creating car objects
car1 = Car("Corolla", 3000)
car2 = Car("Civic", 4000)

# Display details
car1.display_info()
car2.display_info()

# Calculate rent
car1.calculate_rent(3)
car2.calculate_rent(5)
```
