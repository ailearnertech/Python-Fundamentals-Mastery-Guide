# 🔷 CHAPTER 1: INHERITANCE 
## 1️⃣ Concept of Inheritance

#### Inheritance is one of the four pillars of OOP.

#### Inheritance allows one class to acquire properties and methods of another class.

#### In simple words:

#### A new class can reuse code from an existing class.

## 2️⃣ Why Inheritance is Used?

#### Inheritance is used for:

#### ✔ Code Reusability
#### ✔ Reducing Redundancy
#### ✔ Logical Hierarchy
#### ✔ Extensibility
#### ✔ Real-world Modeling

## 3️⃣ Real-World Understanding

### Example:

#### Parent → Human
#### Child → Student

#### Student is a Human.
#### So Student can inherit features of Human.

#### Human has:

#### • name

#### • age

#### • eat()

#### • sleep()

#### Student can reuse all these without rewriting.

## 4️⃣ Basic Terminology

| Term              | Meaning                  |
| ----------------- | ------------------------ |
| Parent Class      | Base Class / Superclass  |
| Child Class       | Derived Class / Subclass |
| Inherit           | Acquiring features       |
| IS-A Relationship | Logical relation         |

## 5️⃣ Syntax of Inheritance in Python
#### class Parent:
####     # parent code

#### class Child(Parent):
####     # child code

#### Child class name followed by parentheses containing parent class name.

## 6️⃣ Single Inheritance

#### Single inheritance means:

#### One child inherits from one parent.

### Example

```python
# Parent class
class Person:
    
    def __init__(self, name):
        self.name = name
    
    def display(self):
        print("Name:", self.name)


# Child class inheriting Person
class Student(Person):
    
    def study(self):
        print(self.name, "is studying.")


# Creating object
student1 = Student("Ali")

student1.display()   # Inherited method
student1.study()     # Own method
```

#### What Happened?

####  Student inherited:

#### ✔ init
#### ✔ display()

#### Without rewriting.

## 7️⃣ Constructor in Inheritance

### Important Rule:

#### Child class does NOT automatically inherit parent constructor in full behavior when defining its own constructor.

#### Let’s understand properly.

### Case 1: Child Without Constructor

```python
class Person:
    
    def __init__(self, name):
        self.name = name


class Student(Person):
    pass


student1 = Student("Ali")
print(student1.name)
```

### Case 2: Child Has Its Own Constructor

```python
class Person:
    
    def __init__(self, name):
        self.name = name


class Student(Person):
    
    def __init__(self, name, grade):
        self.grade = grade


student1 = Student("Ali", "A")
```

## 8️⃣ super() Function

#### To call parent constructor:

#### super().__init__(arguments)
### Correct Version

```python
class Person:
    
    def __init__(self, name):
        self.name = name


class Student(Person):
    
    def __init__(self, name, grade):
        super().__init__(name)  # Calling parent constructor
        self.grade = grade


student1 = Student("Ali", "A")

print(student1.name)
print(student1.grade)
```

#### super() ensures proper initialization.

## 9️⃣ Method Overriding

#### Child class can modify parent method.

#### This is called Method Overriding.

### Example

```python
class Animal:
    
    def sound(self):
        print("Animal makes a sound")


class Dog(Animal):
    
    def sound(self):  # Overriding
        print("Dog barks")


dog1 = Dog()

dog1.sound()
```

## 🔟 Calling Parent Method After Overriding

```python
class Animal:
    
    def sound(self):
        print("Animal makes a sound")


class Dog(Animal):
    
    def sound(self):
        super().sound()
        print("Dog barks")


dog1 = Dog()
dog1.sound()
```

## 1️⃣1️⃣ Types of Inheritance in Python

#### Python supports 5 types:

#### 1. Single

#### 2. Multiple

#### 3. Multilevel

#### 4. Hierarchical

#### 5. Hybrid

## 1️⃣2️⃣ Multiple Inheritance

#### Child inherits from multiple parents.

```python
class Father:
    def skills(self):
        print("Gardening")


class Mother:
    def talent(self):
        print("Cooking")


class Child(Father, Mother):
    pass


child1 = Child()

child1.skills()
child1.talent()
```

## 1️⃣3️⃣ Multilevel Inheritance

#### Inheritance chain.

#### Grandparent → Parent → Child

```python
class Grandparent:
    def house(self):
        print("Owns a house")


class Parent(Grandparent):
    def car(self):
        print("Owns a car")


class Child(Parent):
    def bike(self):
        print("Owns a bike")


child1 = Child()

child1.house()
child1.car()
child1.bike()
```

## 1️⃣4️⃣ Hierarchical Inheritance

#### One parent → Multiple children

```python

class Person:
    def speak(self):
        print("Speaking")


class Student(Person):
    pass


class Teacher(Person):
    pass


student1 = Student()
teacher1 = Teacher()

student1.speak()
teacher1.speak()
```

## 1️⃣5️⃣ Hybrid Inheritance

#### Combination of multiple types.

#### Python allows it.

### Example:

#### Multiple + Multilevel combined.

## 1️⃣6️⃣ Method Resolution Order (MRO)

#### Important in Multiple Inheritance.

### MRO decides:

#### Which parent method gets called first.

#### Check using:

#### print(ClassName.__mro__)

#### or

#### help(ClassName)

#### Python follows:

#### C3 Linearization Algorithm.

#### Order: Left to Right.

#### ✔ Maintainability
#### ✔ Extensibility
#### ✔ Logical Structure
#### ✔ Reduced Code Duplication

## 1️⃣8️⃣ Disadvantages of Inheritance

#### ❌ Tight Coupling
#### ❌ Complex Hierarchy
#### ❌ Difficult Debugging in Multiple Inheritance
#### ❌ Diamond Problem (Handled by MRO in Python)

## 1️⃣9️⃣ Real-World Complete Example

```python
class Employee:
    
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
    
    def show_details(self):
        print("Name:", self.name)
        print("Salary:", self.salary)


class Manager(Employee):
    
    def __init__(self, name, salary, department):
        super().__init__(name, salary)
        self.department = department
    
    def show_details(self):
        super().show_details()
        print("Department:", self.department)


manager1 = Manager("Ali", 80000, "IT")

manager1.show_details()
```

# 🔷 CHAPTER 2: POLYMORPHISM 

## 1️⃣ Concept of Polymorphism

#### Polymorphism means:

#### One name, many forms.

#### In simple words:

#### Same method name
#### But different behavior.

## 2️⃣ Real-World Understanding

### Example:

#### A person can:

#### • Act as Teacher in school

#### • Act as Father at home

#### • Act as Customer in shop

#### Same person
#### Different roles.

#### That is polymorphism.

## 3️⃣ Why Polymorphism is Important?

#### ✔ Flexibility
#### ✔ Code Reusability
#### ✔ Dynamic Behavior
#### ✔ Clean Design
#### ✔ Scalable Applications

## 4️⃣ Types of Polymorphism in Python

#### Python mainly supports:

### 1️⃣ Compile-Time Polymorphism (Simulated)
### 2️⃣ Runtime Polymorphism

## 5️⃣ Compile-Time Polymorphism (Method Overloading Concept)

#### Python does NOT support traditional method overloading like C++.

#### In C++:

#### int add(int a, int b);
#### int add(int a, int b, int c);

#### But in Python:

#### If we define same method twice → Last one overrides.

### Example:

```python
class Demo:
    
    def add(self, a, b):
        return a + b
    
    def add(self, a, b, c):  # This replaces previous one
        return a + b + c
```

#### Only second add() exists.

## 6️⃣ How Python Simulates Method Overloading

### Using Default Arguments:

```python

class Calculator:
    
    def add(self, a, b, c=0):  # c has default value
        return a + b + c


calc = Calculator()

print(calc.add(2, 3))      # Uses default c=0
print(calc.add(2, 3, 4))   # Uses provided value
```

#### This is flexible behavior.

## 7️⃣ Runtime Polymorphism (Method Overriding)

#### Most important type in Python.

#### Child class overrides parent method.

### Example

```python
class Animal:
    
    def sound(self):
        print("Animal makes a sound")


class Dog(Animal):
    
    def sound(self):  # Overriding parent method
        print("Dog barks")


class Cat(Animal):
    
    def sound(self):  # Overriding parent method
        print("Cat meows")


dog1 = Dog()
cat1 = Cat()

dog1.sound()
cat1.sound()
```

#### Same method name → Different behavior.

## 8️⃣ Polymorphism Using Common Interface

```python
class Bird:
    
    def fly(self):
        print("Bird is flying")


class Airplane:
    
    def fly(self):
        print("Airplane is flying")


def start_flying(entity):
    entity.fly()   # Works for both


bird = Bird()
plane = Airplane()

start_flying(bird)
start_flying(plane)
```

#### Here:

#### Python doesn’t care about class type.
#### It only checks if method exists.

#### This is powerful polymorphism.

## 9️⃣ Duck Typing

#### Very important concept.

#### If it looks like a duck and quacks like a duck, it is a duck.

### Meaning:

#### 1. Object type doesn’t matter.
#### 2. Method existence matters.

### Example:

```python
class Student:
    
    def speak(self):
        print("Student speaking")


class Teacher:
    
    def speak(self):
        print("Teacher speaking")


def start_speaking(person):
    person.speak()


s = Student()
t = Teacher()

start_speaking(s)
start_speaking(t)
```

#### Both work.

#### Python checks behavior, not type.

## 🔟 Operator Overloading

#### Python allows operator polymorphism.

### Example:

```python
print(5 + 3)        # Adds numbers
print("A" + "B")    # Concatenates strings
```

#### Same operator +
#### Different behavior.

#### Custom Operator Overloading

```python
class Number:
    
    def __init__(self, value):
        self.value = value
    
    def __add__(self, other):
        return self.value + other.value


num1 = Number(10)
num2 = Number(20)

print(num1 + num2)
```

#### Here:

#### + works for custom objects.

## 1️⃣1️⃣ Method Resolution in Polymorphism

#### When calling method:

#### Python searches:

#### 1. Current class

#### 2. Parent class

#### 3. According to MRO

## 1️⃣2️⃣ Advantages of Polymorphism

#### ✔ Clean code
#### ✔ Flexible design
#### ✔ Easy maintenance
#### ✔ Extensible programs
#### ✔ Reduced conditional statements

## 1️⃣3️⃣ Disadvantages

#### ❌ Can make debugging harder
#### ❌ Complex in large hierarchies
#### ❌ Improper design causes confusion

## 1️⃣4️⃣ Real-World Case Study

### Payment System Example.

```python
class Payment:
    
    def pay(self, amount):
        print("Processing payment")


class CreditCard(Payment):
    
    def pay(self, amount):
        print("Paid", amount, "using Credit Card")


class PayPal(Payment):
    
    def pay(self, amount):
        print("Paid", amount, "using PayPal")


def process_payment(payment_method, amount):
    payment_method.pay(amount)


cc = CreditCard()
pp = PayPal()

process_payment(cc, 500)
process_payment(pp, 1000)
```

## 1️⃣5️⃣ Difference Between Overloading & Overriding

| Feature        | Overloading                       | Overriding                  |
| -------------- | --------------------------------- | --------------------------- |
| Definition     | Same method, different parameters | Child changes parent method |
| Python Support | Simulated                         | Fully Supported             |
| Occurs         | Same class                        | Parent & Child              |
| Time           | Compile-time concept              | Runtime                     |

# 🔷 CHAPTER 3: ENCAPSULATION 
## 1️⃣ Concept of Encapsulation

#### Encapsulation means:

#### Wrapping data (variables) and methods (functions) together inside a class
#### AND
#### Restricting direct access to some data.

#### In simple words:

#### • Hide internal details
#### • Control access to data

## 2️⃣ Why Encapsulation is Important?

#### ✔ Data Protection
#### ✔ Security
#### ✔ Controlled Access
#### ✔ Prevents accidental modification
#### ✔ Clean structure

## 3️⃣ Real-World Example

#### Think about ATM Machine.

#### You:

#### • Insert card

#### • Enter PIN

#### • Withdraw money

#### But you CANNOT:

#### • Directly access bank database

#### • Directly change balance

#### System hides internal data.

#### That is encapsulation.

## 4️⃣ Encapsulation in Python

#### In Python, encapsulation is achieved using:

#### 1️⃣ Public Members
#### 2️⃣ Protected Members
#### 3️⃣ Private Members

## 5️⃣ Public Members

#### Public members are accessible everywhere.

#### By default, all variables in Python are public.

### Example

```python
class Student:
    
    def __init__(self, name):
        self.name = name   # Public variable
    
    def display(self):
        print("Name:", self.name)


student1 = Student("Ali")

print(student1.name)  # Accessible
student1.display()
```

## 6️⃣ Protected Members

#### Protected members use single underscore _.

### Convention:

#### Should not be accessed outside class
#### But technically accessible.

### Example

```python
class Student:
    
    def __init__(self, name):
        self._name = name   # Protected variable
    
    def display(self):
        print("Name:", self._name)


student1 = Student("Ali")

print(student1._name)  # Still accessible (not strictly restricted)
```

#### Important:

#### Single underscore is just a warning for developers.

## 7️⃣ Private Members

#### Private members use double underscore __.

#### Python performs name mangling.

### Example

```python
class Student:
    
    def __init__(self, name):
        self.__name = name   # Private variable
    
    def display(self):
        print("Name:", self.__name)


student1 = Student("Ali")

student1.display()

# print(student1.__name)  # This will cause error
```

## 8️⃣ Name Mangling in Python

#### Python internally changes:

#### self.__name

#### to:

#### _ClassName__name

### Example:

```python
print(student1._Student__name)
```

#### It works.

#### So private is not absolute security
#### But strong protection mechanism.

## 9️⃣ Data Hiding Concept

#### Encapsulation allows:

#### 1. Hide internal data
#### 2. Allow access through methods

#### This improves security.

## 🔟 Getter and Setter Methods

#### To access private variables safely:

#### We use getters and setters.

### Example

```python
class BankAccount:
    
    def __init__(self, balance):
        self.__balance = balance   # Private variable
    
    # Getter method
    def get_balance(self):
        return self.__balance
    
    # Setter method
    def set_balance(self, amount):
        if amount >= 0:
            self.__balance = amount
        else:
            print("Invalid amount")


account = BankAccount(5000)

print(account.get_balance())  # Access through getter

account.set_balance(7000)
print(account.get_balance())
```

## 1️⃣1️⃣ Encapsulation with Validation

```python
class Student:
    
    def __init__(self, marks):
        self.__marks = marks
    
    def set_marks(self, marks):
        if 0 <= marks <= 100:
            self.__marks = marks
        else:
            print("Invalid marks")
    
    def get_marks(self):
        return self.__marks


s = Student(85)

s.set_marks(120)   # Invalid
print(s.get_marks())
```

#### Encapsulation protects data integrity.

## 1️⃣2️⃣ Encapsulation in Inheritance

#### Private members are NOT directly accessible in child class.

### Example:

```python
class Parent:
    
    def __init__(self):
        self.__value = 10


class Child(Parent):
    
    def show(self):
        # print(self.__value)  # Error
        pass
```

## 1️⃣3️⃣ Advantages of Encapsulation

#### ✔ Better security
#### ✔ Data hiding
#### ✔ Controlled modification
#### ✔ Code flexibility
#### ✔ Easy maintenance

## 1️⃣4️⃣ Disadvantages

#### ❌ Slightly more code
#### ❌ Can increase complexity
#### ❌ Not absolute security

## 1️⃣5️⃣ Public vs Protected vs Private Comparison

| Type      | Symbol | Accessible Outside    | Accessible in Child |
| --------- | ------ | --------------------- | ------------------- |
| Public    | None   | Yes                   | Yes                 |
| Protected | _      | Yes (not recommended) | Yes                 |
| Private   | __     | No (directly)         | No                  |

## 1️⃣6️⃣ Real-World Case Study

### ATM System Example

```python
class ATM:
    
    def __init__(self, balance):
        self.__balance = balance
    
    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print("Withdraw successful")
        else:
            print("Insufficient balance")
    
    def check_balance(self):
        print("Balance:", self.__balance)


atm = ATM(10000)

atm.withdraw(2000)
atm.check_balance()

# Direct access not allowed
# print(atm.__balance)  # Error
```

# 🔷 CHAPTER 4: ABSTRACTION 

## 1️⃣ Concept of Abstraction

#### Abstraction means:

#### Hiding internal implementation details
#### Showing only essential features to the user.

#### In simple words:

#### User sees what the object does
#### User does NOT see how it does it

## 2️⃣ Real-World Example

#### Think about a Car.

#### You:

#### • Start engine

#### • Press accelerator

#### • Apply brakes

####  But you don’t know:

#### • Internal engine mechanism

#### • Fuel injection process

#### • Gear rotation logic

#### • System hides complexity.

#### That is abstraction.

## 3️⃣ Why Abstraction is Important?

#### ✔ Reduces complexity
#### ✔ Improves security
#### ✔ Makes code cleaner
#### ✔ Improves maintainability
#### ✔ Forces structured design

## 4️⃣ Abstraction in Python

#### Python achieves abstraction using:

#### 1️⃣ Abstract Classes
#### 2️⃣ Abstract Methods
#### 3️⃣ abc Module

## 5️⃣ What is an Abstract Class?

#### An abstract class:

#### • Cannot be instantiated directly

#### • Contains at least one abstract method

#### Abstract method:

#### • Declared but not implemented

#### • Must be implemented by child class

## 6️⃣ abc Module

#### Python provides:

#### from abc import ABC, abstractmethod

#### ABC → Abstract Base Class

#### @abstractmethod → Decorator

## 7️⃣ Basic Abstract Class Example

```python
from abc import ABC, abstractmethod

# Abstract class
class Shape(ABC):
    
    @abstractmethod
    def area(self):
        pass  # No implementation


# Child class
class Circle(Shape):
    
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return 3.14 * self.radius * self.radius


circle1 = Circle(5)
print(circle1.area())
```

### Important Rule

#### If child class does NOT implement abstract method:

#### It cannot be instantiated.

### Example:

```python

class Rectangle(Shape):
    pass

rect = Rectangle()  # Error
```

#### Python raises TypeError.

## 8️⃣ Why We Cannot Create Object of Abstract Class?

#### Because it is incomplete.

### Example:

```python
shape = Shape()  # Error
```

#### Abstract class only defines blueprint.

## 9️⃣ Abstraction vs Encapsulation

#### Very Important Difference:

| Encapsulation          | Abstraction               |
| ---------------------- | ------------------------- |
| Hides data             | Hides implementation      |
| Uses private variables | Uses abstract methods     |
| Focus on security      | Focus on design structure |


## 🔟 Real-World Case Study – Payment System

```python
from abc import ABC, abstractmethod

class Payment(ABC):
    
    @abstractmethod
    def pay(self, amount):
        pass


class CreditCard(Payment):
    
    def pay(self, amount):
        print("Paid", amount, "using Credit Card")


class PayPal(Payment):
    
    def pay(self, amount):
        print("Paid", amount, "using PayPal")


cc = CreditCard()
pp = PayPal()

cc.pay(500)
pp.pay(1000)
```

#### User only sees:

#### pay()

#### Not internal payment processing logic.

## 1️⃣1️⃣ Multiple Abstract Methods

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    
    @abstractmethod
    def start(self):
        pass
    
    @abstractmethod
    def stop(self):
        pass


class Car(Vehicle):
    
    def start(self):
        print("Car started")
    
    def stop(self):
        print("Car stopped")


car1 = Car()
car1.start()
car1.stop()
```

#### Child must implement ALL abstract methods.

## 1️⃣2️⃣ Abstract Class With Normal Methods

#### Abstract class can have:

#### ✔ Abstract methods
#### ✔ Normal methods

### Example:

```python
from abc import ABC, abstractmethod

class Employee(ABC):
    
    def show_role(self):
        print("This is an employee")
    
    @abstractmethod
    def calculate_salary(self):
        pass


class Manager(Employee):
    
    def calculate_salary(self):
        print("Salary calculated for manager")


manager1 = Manager()
manager1.show_role()
manager1.calculate_salary()
```

## 1️⃣3️⃣ Interface-Like Behavior in Python

#### Python does not have separate interface keyword.

#### Abstract class acts like interface.

#### 1️⃣4️⃣ Advantages of Abstraction

#### ✔ Clean structure
#### ✔ Forces implementation
#### ✔ Reduces errors
#### ✔ Improves scalability
#### ✔ Encourages proper design

## 1️⃣5️⃣ Disadvantages

#### ❌ Slightly complex for beginners
#### ❌ More code required
#### ❌ Overuse can make design rigid

## 1️⃣6️⃣ Complete Real-World Example – Online Shopping


```python
from abc import ABC, abstractmethod

class Product(ABC):
    
    @abstractmethod
    def get_price(self):
        pass


class Book(Product):
    
    def __init__(self, price):
        self.price = price
    
    def get_price(self):
        return self.price


class Laptop(Product):
    
    def __init__(self, price):
        self.price = price
    
    def get_price(self):
        return self.price


book1 = Book(500)
laptop1 = Laptop(80000)

print(book1.get_price())
print(laptop1.get_price())
```
