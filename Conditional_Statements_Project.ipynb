# -> Decision Making – Introduction
## What is Decision Making in Programming?

#### Decision making in programming is the process by which a program chooses which action to take based on certain conditions. 
#### Just like humans make decisions in real life, computers can also “decide” what to do if we give them rules.
#### In programming, conditional statements allow a computer to take actions based on True or False conditions.

## Real-Life Examples of Decision Making:

| Situation                 | Decision / Action |
| ------------------------- | ----------------- |
| It is raining             | Take an umbrella  |
| Temperature is above 30°C | Turn on the fan   |
| Exam marks ≥ 50%          | Pass the exam     |
| Exam marks < 50%          | Fail the exam     |


## Why Decision Making is Important?

#### • Without decision making, programs can only run line by line in a linear fashion.

#### • With decision making, programs can respond dynamically to different inputs or situations.

## Examples in real-world programming:

#### • Login systems (check username & password)

#### • Online shopping (apply discount if coupon is valid)

#### • Games (player health < 0 → game over)

#### • Banking (withdraw amount ≤ balance → allow withdrawal)

## How Decision Making Works?

#### 1.Condition Evaluation: Check if a statement is True or False.

#### 2.Action: If True → execute a specific block of code; if False → execute alternative block (if defined).

#### 3.Flow Control: Program changes the execution path based on conditions.


## Flowchart Representation:


#### Start  
#### &nbsp;&nbsp;&nbsp;&nbsp;↓  
#### Check Condition  
#### &nbsp;&nbsp;&nbsp;&nbsp;↓  
#### If True → Execute True Block  
#### &nbsp;&nbsp;&nbsp;&nbsp;↓  
#### If False → Execute False Block  
#### &nbsp;&nbsp;&nbsp;&nbsp;↓  
#### End


# -> Conditional Statements – Overview

## 1️⃣ What are Conditional Statements?

#### Conditional statements are special programming statements that allow a program to make decisions based on certain conditions.

#### In simple words:

#### A conditional statement tells the computer:
#### “If this condition is true, do this. Otherwise, do something else.”

#### A condition is always checked as True or False.

## 2️⃣ Why Do We Need Conditional Statements?

#### By default, a program runs line by line from top to bottom.
#### This is called sequential execution.
#### Example:

```python
print("Step 1")
print("Step 2")
print("Step 3")
```

#### The program has no decision-making power here.

#### But in real-world applications, programs must:

#### • Validate user input

#### • Check login credentials

#### • Compare numbers

#### • Make logical decisions

#### • Control program flow

#### Without conditional statements, programs would be:

#### • Static

#### • Unintelligent

#### • Non-interactive

## 3️⃣ How Conditional Statements Work?

#### Every conditional statement follows three steps:

| Step | Description                             |
| ---- | --------------------------------------- |
| 1    | Evaluate a condition                    |
| 2    | Check if result is True or False        |
| 3    | Execute the corresponding block of code |


## Flow Representation:

#### Start  
#### &nbsp;&nbsp;&nbsp;&nbsp;|  
#### Check Condition  
#### &nbsp;&nbsp;&nbsp;&nbsp;|  
####  True --------> Execute Block A
#### &nbsp;&nbsp;&nbsp;&nbsp;| 
####  False -------> Execute Block B
#### &nbsp;&nbsp;&nbsp;&nbsp;| 
#### End

#### This structure allows the program to change direction.

## 4️⃣ Core Components of Conditional Statements:

#### To understand conditional statements properly, we must understand their building blocks:

### 🔹 1. Condition

#### A logical expression that evaluates to True or False.

#### Example:

 #### age >= 18

### 🔹 2. Boolean Result

#### The condition always produces:

#### True
#### False

### 🔹 3. Block of Code

#### The code that executes if the condition is satisfied.

#### In Python, indentation defines the block:

#### if age >= 18:
  ####   print("Adult")

## 5️⃣ Types of Conditional Statements in Python:

#### Python provides several types:

| Type                     | Purpose                            |
| ------------------------ | ---------------------------------- |
| if                       | Execute code if condition is True  |
| if-else                  | Choose between two options         |
| if-elif-else             | Check multiple conditions          |
| Nested if                | Condition inside another condition |
| Conditional inside loops | Decision during repetition         |


#### We will study each type in detail in the next sections.

# -> Boolean Data Type

## 1️⃣ What is Boolean Data Type?
#### In programming, a Boolean is a data type that has only two possible values:

#### True  or  False

#### • Named after George Boole, an English mathematician

#### • Boolean values are the foundation of decision-making in programming

## 2️⃣ Why Boolean is Important?

#### • Conditional statements always evaluate to True or False

#### • Helps the program decide which path to take

#### • Used in comparison operations, logical operations, loops, and decision-making

#### Example in real life:

| Situation      | Boolean Representation |
| -------------- | ---------------------- |
| Door is open   | True                   |
| Door is closed | False                  |
| Light is on    | True                   |
| Light is off   | False                  |


## 3️⃣ Boolean Values in Python:

#### Python has two boolean constants:

```python
 x = True
 y = False

 print(x)     # True
 print(y)     # False
```

#### • Case-sensitive → True and False must have capital first letter

#### • Lowercase true or false will throw error

## 4️⃣ Boolean Expressions:

#### A Boolean expression is a statement that evaluates to True or False.

### Example:

```python
 a = 10
 b = 5

 print(a > b)   # True
 print(a < b)   # False
 print(a == b)  # False
 print(a != b)  # True
```

## 5️⃣ Boolean with Conditional Statements:

#### Boolean expressions are used in if, if-else, and loops.

#### • age >= 18 → evaluates to True or False

#### • The program decides the action based on this Boolean value

## 6️⃣ Boolean Operators in Python:

#### Boolean values can be combined using logical operators:

| Operator | Meaning          | Example          | Result |
| -------- | ---------------- | ---------------- | ------ |
| `and`    | Both True → True | `True and False` | False  |
| `or`     | Any True → True  | `True or False`  | True   |
| `not`    | Reverse value    | `not True`       | False  |


#### Example in code:

```python
 a = True
 b = False

 print(a and b)  # False
 print(a or b)   # True
 print(not a)    # False
```

## 7️⃣ Truthy and Falsy Values in Python:

#### In Python, not only True/False but certain values behave like boolean:

| Value                                 | Boolean Equivalent |
| ------------------------------------- | ------------------ |
| `0`                                   | False              |
| `""`                                  | False              |
| `[]`                                  | False              |
| `{}`                                  | False              |
| `None`                                | False              |
| All other non-zero / non-empty values | True               |


### Example:

```python
 x = 0
 y = 5

 if x:
     print("X is True")
 else:
     print("X is False")  # X is False

 if y:
    print("Y is True")   # Y is True
```

#### • Zero, empty container, or None → False

#### • Non-zero, non-empty → True

# -> Comparison Operators in Python

## 1️⃣ What are Comparison Operators?

#### Comparison operators are used to compare two values.
#### The result of a comparison is always a Boolean value: True or False.

#### "These operators are the foundation of decision making because they allow the program to evaluate conditions."

## 2️⃣ Why We Use Comparison Operators?

#### • To check equality or difference between values

#### • To determine greater than or less than relationships

#### • To control program flow using conditional statements

## 3️⃣ List of Comparison Operators in Python:

| Operator | Meaning               | Example  | Output |
| -------- | --------------------- | -------- | ------ |
| `==`     | Equal to              | `5 == 5` | True   |
| `!=`     | Not equal             | `5 != 3` | True   |
| `>`      | Greater than          | `10 > 3` | True   |
| `<`      | Less than             | `2 < 7`  | True   |
| `>=`     | Greater than or equal | `5 >= 5` | True   |
| `<=`     | Less than or equal    | `4 <= 6` | True   |


## 4️⃣ Example in Python:

```python
 a = 10
 b = 5

 print(a == b)  # False
 print(a != b)  # True
 print(a > b)   # True# print(a < b) 
 False
 print(a >= 10) # True
 print(b <= 5)  # True
```

## 5️⃣ Using Comparison Operators in Conditional Statements:

#### Comparison operators are usually used inside if / if-else statements.

```python
marks = 75

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
else:
    print("Fail")
```

## 6️⃣ Combining Comparison Operators with Logical Operators:

#### You can combine multiple comparisons using and / or / not.

```python
age = 20
citizen = True

if age >= 18 and citizen:
    print("Eligible to vote")
else:
    print("Not eligible")
```

# -> Logical Operators in Python

## 1️⃣ What are Logical Operators?

#### Logical operators are used to combine multiple conditions and make decisions more complex and intelligent.

#### They work with Boolean values (True or False) and return True or False as the result.

## 2️⃣ Why Logical Operators are Important

#### • Sometimes a single condition is not enough

#### • Programs need to check multiple conditions at once

#### • Logical operators help in writing concise and readable decision logic

## 3️⃣ Types of Logical Operators in Python?

| Operator | Name | Meaning                                    | Example          | Result |
| -------- | ---- | ------------------------------------------ | ---------------- | ------ |
| `and`    | AND  | True if **both** conditions are True       | `True and False` | False  |
| `or`     | OR   | True if **at least one** condition is True | `True or False`  | True   |
| `not`    | NOT  | Reverse the Boolean value                  | `not True`       | False  |


## 4️⃣ Examples of Logical Operators:
### 🔹 AND Operator

```python
age = 20
citizen = True

if age >= 18 and citizen:
    print("Eligible to vote")
else:
    print("Not eligible")

```

#### • Both conditions must be True for eligibility

#### • and → True only if both True

### 🔹 OR Operator

```python
marks = 75
income = 40000

if marks >= 80 or income < 50000:
    print("Eligible for scholarship")
else:
    print("Not eligible")
```

#### • At least one condition True → block executes

#### • or → True if any one condition True

### 🔹 NOT Operator

```python
is_raining = False

if not is_raining:
    print("Go outside")
else:
    print("Take an umbrella")
```

#### • not reverses the Boolean value

#### • Here, not False → True → "Go outside"

## 5️⃣ Combining Multiple Logical Operators:

#### You can combine multiple operators for complex conditions:

```python
age = 20
citizen = True
income = 40000

if age >= 18 and citizen and income < 50000:
    print("Eligible for special program")
```

#### •  All 3 conditions must be True

#### •  Logical operators can be chained in one line

# -> Simple if Statement in Python

## 1️⃣ What is a Simple if Statement?

#### A simple if statement allows a program to execute a block of code only if a certain condition is True.

#### It is the most basic form of decision-making in programming.

## 2️⃣ Syntax of Simple if Statement:

#### if condition:
#### # Block of code to execute

## 3️⃣ Flowchart Representation:

#### Start  
#### &nbsp;&nbsp;&nbsp;&nbsp;|  
#### Check Condition  
#### &nbsp;&nbsp;&nbsp;&nbsp;|  
####  True --------> Execute code
#### &nbsp;&nbsp;&nbsp;&nbsp;| 
####  False -------> Skip code
#### &nbsp;&nbsp;&nbsp;&nbsp;| 
#### End

####  • Simple, linear decision

####  • No alternative path defined

## 4️⃣ Example 1 – Positive Number Check

```python
number = 10

if number > 0:
    print("Positive Number")
```

## 5️⃣ Example 2 – Passing Marks

```python
marks = 65

if marks >= 50:
    print("You Passed!")
```

## 6️⃣ Important Points About Simple if

#### • Used when only one condition matters

#### • Executes code only when condition is True

#### • Can have multiple simple if statements independently:
#### • Python checks all conditions separately

```python
marks = 75

if marks >= 90:
    print("Grade A")
if marks >= 75:
    print("Grade B")
if marks >= 50:
    print("Pass")
```

# -> if-else Statement in Python

## 1️⃣ What is an if-else Statement?

#### The if-else statement allows a program to choose between two paths:

#### • Execute one block of code if a condition is True

#### • Execute another block if the condition is False

#### Unlike simple if, which handles only True, if-else provides an alternative path for False.

## 2️⃣ Syntax of if-else Statement:

#### if condition:
  ####  # Block of code if condition is True
#### else:
 ####   # Block of code if condition is False


### Rules:

#### 1. The if block executes if the condition is True

#### 2. The else block executes if the condition is False

#### 3. Both blocks must be indented properly in Python

## 3️⃣ Flowchart Representation:

#### Start  
#### &nbsp;&nbsp;&nbsp;&nbsp;|  
#### Check Condition  
#### &nbsp;&nbsp;&nbsp;&nbsp;|  
####  True --------> Execute if Block
#### &nbsp;&nbsp;&nbsp;&nbsp;| 
####  False -------> Execute else Block
#### &nbsp;&nbsp;&nbsp;&nbsp;| 
#### End

## 4️⃣ Example 1 – Check Passing Marks

```python
marks = 45

if marks >= 50:
    print("You Passed")
else:
    print("You Failed")
```

## 5️⃣ Example 2 – Even or Odd Number

```python
number = 7

if number % 2 == 0:
    print("Even Number")
else:
    print("Odd Number")
```

## 6️⃣ Important Points About if-else

#### • Handles both True and False cases

#### • Always exactly one block executes

#### • Can be nested inside loops or other if statements

## 7️⃣ Multiple Independent if-else Statements

```python
marks = 75

if marks >= 50:
    print("Pass")
else:
    print("Fail")

if marks >= 75:
    print("Distinction")
else:
    print("No Distinction")
```

#### • Each if-else is evaluated independently

#### • Python does not stop after first block

# -> if-elif-else Statement in Python

## 1️⃣ What is an if-elif-else Statement?

#### The if-elif-else statement allows a program to check multiple conditions in order and execute the block corresponding to the first condition that is True.

#### • if → First condition

#### • elif → Additional conditions (short for "else if")

#### • else → Default block if none of the conditions are True

#### This is used when there are more than two possible choices.

## 2️⃣ Syntax of if-elif-else Statement:

#### if condition1:
 ####    # Block executed if condition1 is True
#### elif condition2:
####     # Block executed if condition2 is True
#### elif condition3:
####     # Block executed if condition3 is True
#### else:
####     # Block executed if none of the above conditions are True


### Rules:

#### 1. At least one if is required

#### 2. Zero or more elif blocks can be used

#### 3. Optional else block can handle default case

#### 4. Only the first True block executes; the rest are skipped

## 3️⃣ Flowchart Representation

#### Start  
#### &nbsp;&nbsp;&nbsp;&nbsp;|  
#### Check Condition 1  
#### &nbsp;&nbsp;&nbsp;&nbsp;|  
####  True --------> Execute Block 1
#### &nbsp;&nbsp;&nbsp;&nbsp;| 
####  False -------> Check Condition 2
#### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;| 
####  True --------> Execute Block 2
####  False -------> Check Condition 3
#### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;|
####  True --------> Execute Block 3
####  False -------> Execute else Block
#### &nbsp;&nbsp;&nbsp;&nbsp;| 
#### End

#### • Shows multiple conditions checked sequentially

## 4️⃣ Example 1 – Grading System

```python
marks = 85

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 60:
    print("Grade C")
else:
    print("Fail")
```

## 5️⃣ Example 2 – Ticket Pricing

```python
age = 12

if age <= 5:
    print("Free Ticket")
elif age <= 12:
    print("Child Ticket")
elif age <= 60:
    print("Adult Ticket")
else:
    print("Senior Citizen Ticket")
```

## 6️⃣ Important Points About if-elif-else

#### 1. Used when more than two choices exist

#### 2. Only one block executes

#### 3. Order of conditions matters

#### 4. Can be nested inside loops or other conditionals

# -> Nested if Statement in Python

## 1️⃣ What is a Nested if Statement?

#### A nested if statement is an if or if-else statement inside another if or if-else statement.

#### This is used when a decision depends on another decision.

#### • Allows multi-level decision-making

#### • Makes programs more specific and flexible

## 2️⃣ Syntax of Nested if:

#### if condition1:
####     # Executes if condition1 is True
 ####    if condition2:
 ####        # Executes if condition2 is also True
 ####    else:
 ####        # Executes if condition2 is False
#### else:
####     # Executes if condition1 is False


### Rules:

#### 1. Inner if is executed only if the outer if is True

#### 2. Indentation is critical in Python

#### 3.Can be nested multiple levels, but avoid excessive nesting

## 3️⃣ Flowchart Representation:

#### Start
#### &nbsp;&nbsp;&nbsp;&nbsp;  |
#### Check Condition1
#### &nbsp;&nbsp;&nbsp;&nbsp;   |
####   |---- True ----> Check Condition2
####   |         &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;|
####   |         &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;|---- True ----> Execute Block2
####   |         &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;|---- False ---> Execute Else Block2
####   |
####   |---- False ---> Execute Else Block1
####   |
#### End


#### • Shows multi-level decision path

## 4️⃣ Example 1 – Number Check

### Problem: Check if a number is positive even, positive odd, or negative

```python
number = 8

if number > 0:
    if number % 2 == 0:
        print("Positive Even Number")
    else:
        print("Positive Odd Number")
else:
    print("Negative Number")
```

## 5️⃣ Example 2 – Eligibility Check

### Problem: Determine eligibility for a special program:

####  • Age ≥ 18

####  • AND Citizenship = True

```python
age = 20
citizen = True

if age >= 18:
    if citizen:
        print("Eligible for Program")
    else:
        print("Not Eligible – Not a Citizen")
else:
    print("Not Eligible – Too Young")
```

## 6️⃣ Important Points About Nested if

#### • Inner if is dependent on outer if

#### • Helps handle complex conditions

#### • Keep nesting manageable to avoid confusion

#### • Can be combined with elif and else
