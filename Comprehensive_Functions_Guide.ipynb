# -> Introduction to Functions

## 1️⃣ What is a Function?

#### A function is:

#### A reusable block of organized code that performs a specific task.

#### Instead of writing the same code again and again, we group it inside a function and reuse it whenever needed.

### Simple Definition

#### A function:

#### • Takes input (optional)

#### • Performs some processing

#### • Returns output (optional)

## 2️⃣ Why Functions Are Needed?

#### Without functions:

#### • Code becomes repetitive

#### • Code becomes lengthy

#### • Debugging becomes difficult

#### • Updating logic becomes hard

#### • Program structure becomes messy

#### • Functions solve all these problems.

## 3️⃣ Problem Without Functions

#### Let’s see an example.

### Without Functions

```python
# Calculate area of rectangle 1
length = 5
width = 3
area = length * width
print("Area:", area)

# Calculate area of rectangle 2
length = 10
width = 4
area = length * width
print("Area:", area)

# Calculate area of rectangle 3
length = 7
width = 6
area = length * width
print("Area:", area)
```

## 4️⃣ Solution Using Function

#### Now we organize the logic inside a function.

```python
def calculate_area(length, width):
    area = length * width
    print("Area:", area)

calculate_area(5, 3)
calculate_area(10, 4)
calculate_area(7, 6)
```

#### Similarly:

#### • Function = machine

#### • Parameters = input

#### • Return value = output

## 4️⃣ Recursive Functions
### 🔹 Definition

####  A recursive function is a function that calls itself.

#### Used when:

#### • Problem can be broken into smaller similar problems.

### 🔹 Important Rule

####  Recursive functions must have:

#### 1.  Base case (stopping condition)

#### 2. Recursive case (self-call)

#### Without base case → infinite recursion → error.

### Example 4 – Factorial Using Recursion

```python
def factorial(n):
    if n == 0:          # Base case
        return 1
    else:
        return n * factorial(n - 1)  # Recursive case

print(factorial(5))
```

     factorial(3)
          ↓
      3 * factorial(2)
            ↓
        2 * factorial(1)
              ↓
          1 * factorial(0)
                ↓
            return 1

# -> Function Parameters and Arguments

## 1️⃣ Parameters vs Arguments
### 🔹 Parameter

#### A parameter is a variable listed inside the function definition.

### 🔹 Argument

#### An argument is the actual value passed to the function when calling it.
### Example

```python
def greet(name):   # name → parameter
    print("Hello", name)

greet("Ali")       # "Ali" → argument
```

## 2️⃣ Positional Arguments
### 🔹 Definition

#### Arguments are passed in the same order as parameters.

#### Order matters.

###  Example

```python
def student(name, age):
    print("Name:", name)
    print("Age:", age)

student("Sara", 20)
```

## 3️⃣ Keyword Arguments
### 🔹 Definition

#### Arguments are passed using parameter names.

#### Order does NOT matter.

###  Example

```python
def student(name, age):
    print("Name:", name)
    print("Age:", age)

student(age=20, name="Sara")
```

## 4️⃣ Default Arguments
### 🔹 Definition

#### A default argument has a pre-defined value.

#### If user does not provide value, default is used.

###  Example

```python
def greet(name="Guest"):
    print("Hello", name)

greet("Ali")
greet()
```

## 5️⃣ Arbitrary Positional Arguments (*args)
### 🔹 Definition

#### Used when we don’t know how many positional arguments will be passed.

#### *args collects arguments into a tuple.

###  Example

```python
def add_numbers(*args):
    print("Arguments:", args)
    print("Sum:", sum(args))

add_numbers(1, 2, 3)
add_numbers(10, 20, 30, 40)
```

## 6️⃣ Arbitrary Keyword Arguments (**kwargs)
### 🔹 Definition

#### Used when we don’t know how many keyword arguments will be passed.

#### **kwargs collects them into a dictionary.

###  Example

```python
def student_info(**kwargs):
    print("Data:", kwargs)

student_info(name="Ali", age=21, grade="A")
```

## 7️⃣ Using All Types Together

#### Python allows combination but order must follow rule:

#### Normal parameters

#### *args

#### **kwargs

###  Example

```python
def demo(a, b, *args, **kwargs):
    print("a:", a)
    print("b:", b)
    print("args:", args)
    print("kwargs:", kwargs)

demo(10, 20, 30, 40, name="Ali", age=25)
```

## 8️⃣ Complete Comparison Table

| Type       | Order Matters        | Default Allowed | Stores As             | When Used                              |
| ---------- | -------------------- | --------------- | --------------------- | -------------------------------------- |
| Positional | Yes                  | No              | Individual variables  | Fixed parameters                       |
| Keyword    | No                   | No              | Individual variables  | Clear calling                          |
| Default    | Yes (placement rule) | Yes             | Variable with default | Optional values                        |
| *args      | Yes                  | No              | Tuple                 | Unknown number of positional arguments |
| **kwargs   | No                   | No              | Dictionary            | Unknown number of keyword arguments    |

# -> Variable Scope in Functions (Local, Global & LEGB Rule)

## 1️⃣ What is Scope?
### 🔹 Definition

#### Scope determines:

#### Where a variable can be accessed in a program.

#### In simple words:
#### Scope defines the visibility of a variable.

## 2️⃣ Local Variables
### 🔹 Definition

#### A variable defined inside a function is called a local variable.

#### It can only be used inside that function.

###  Example

```python
def test():
    x = 10   # local variable
    print(x)

test()
print(x)   # ERROR
```

### Why error?

#### Because x exists only inside test().

#### Once function finishes, local variables are destroyed.

## 3️⃣ Global Variables
### 🔹 Definition

#### A variable defined outside any function is called a global variable.

#### It can be accessed anywhere in the program.

###  Example

```python
x = 50   # global variable

def show():
    print(x)

show()
```

## 4️⃣ Modifying Global Variables

#### By default, you cannot modify a global variable inside a function unless you use global keyword.

###  Example (Error)

```python
x = 10

def change():
    x = x + 5   # ERROR

change(
```

#### Python thinks x is local here.

### ✅ Correct Version

```python
x = 10

def change():
    global x
    x = x + 5

change()
print(x)
```

## 5️⃣ Nested Functions

#### Python allows defining a function inside another function.

### Example

```python
def outer():
    x = 10

    def inner():
        print(x)

    inner()

outer()
```

## 6️⃣ Enclosing Scope

#### When a function is inside another function:

#### Outer function variable becomes enclosing variable.

###  Example

```python
def outer():
    message = "Hello"

    def inner():
        print(message)

    inner()

outer()
```

## 7️⃣ LEGB Rule (Very Important)

#### Python searches variables in this order:

### L → Local
### E → Enclosing
### G → Global
### B → Built-in


#### This is called LEGB Rule.

### Example

```python
x = 100

def outer():
    x = 50

    def inner():
        x = 10
        print(x)

    inner()

outer()
```

### If Not in Enclosing

```python
x = 100

def outer():
    def inner():
        print(x)

    inner()

outer()
```

## 8️⃣ Built-in Scope (Last Level)

#### If variable not found anywhere, Python checks built-in scope.

### Example:

```python
print(len([1,2,3]))
```

#### len() comes from Python built-in scope.

## 9️⃣ nonlocal Keyword (Advanced)

#### Used to modify enclosing variable.

###  Example

```python
def outer():
    x = 10

    def inner():
        nonlocal x
        x = 20

    inner()
    print(x)

outer()
```

## Complete Comparison Table

| Type      | Defined Where    | Accessible Where     | Lifetime            |
| --------- | ---------------- | -------------------- | ------------------- |
| Local     | Inside function  | Only inside function | Until function ends |
| Enclosing | Outer function   | Inner function       | Until outer ends    |
| Global    | Outside function | Whole program        | Entire program      |
| Built-in  | Python system    | Everywhere           | Always available    |


# -> Recursive Functions

## 1️⃣ What is Recursion?
### 🔹 Definition

#### A recursive function is a function that calls itself to solve a smaller version of the same problem.

#### It breaks a big problem into smaller subproblems.

### 🔹 Simple Idea

#### Think like this:

#### Problem → Smaller Same Problem → Smaller Same Problem → Base Case

## 2️⃣ Two Mandatory Parts of Recursion

#### Every recursive function must have:

### ✅ 1. Base Case

#### Stopping condition.

### ✅ 2. Recursive Case

#### Function calling itself.

#### Without base case → Infinite recursion → Program crash.

## 3️⃣ Basic Example – Factorial


```python
def factorial(n):
    if n == 0:          # Base case
        return 1
    else:
        return n * factorial(n - 1)   # Recursive case

print(factorial(5))
```

## 4️⃣ Step-by-Step Execution Trace

#### Let’s trace:

### factorial(3)

### Step 1
#### 3 * factorial(2)

### Step 2
#### 2 * factorial(1)

### Step 3
#### 1 * factorial(0)

### Step 4 (Base Case)
#### return 1


#### Now stack unwinds:

#### 1
#### 1 * 1 = 1
#### 2 * 1 = 2
#### 3 * 2 = 6


### Final Answer = 6

## 5️⃣ Internal Working – Call Stack (Very Important)

#### Recursion uses a data structure called:

#### Call Stack

#### Each function call is stored in memory until it finishes.

### Conceptual Stack View for factorial(3)

    | factorial(3) |
    | factorial(2) |
    | factorial(1) |
    | factorial(0) |

## 6️⃣ Another Example – Fibonacci

```python
def fibonacci(n):
    if n <= 1:
        return n
    else:
        return fibonacci(n-1) + fibonacci(n-2)

print(fibonacci(6))
```

## 7️⃣ Types of Recursion
### 1️⃣ Direct Recursion

#### Function calls itself directly.

### Example:

```python
def test():
    test()
```

### 2️⃣ Indirect Recursion

#### Function A calls B, and B calls A.

### Example:

```python
def A(n):
    if n > 0:
        B(n-1)

def B(n):
    if n > 0:
        A(n-1)
```

### 3️⃣ Tail Recursion

#### Recursive call is the last statement.

### Example:

```python
def countdown(n):
    if n == 0:
        return
    print(n)
    return countdown(n-1)
```

## 8️⃣ Recursion vs Loop

#### Same factorial using loop:

```python
def factorial_loop(n):
    result = 1
    for i in range(1, n+1):
        result *= i
    return result
```

## Comparison Table

| Feature             | Recursion            | Loop               |
| ------------------- | -------------------- | ------------------ |
| Uses function calls | Yes                  | No                 |
| Uses stack memory   | Yes                  | No                 |
| Memory usage        | Higher               | Lower              |
| Code length         | Shorter              | Slightly longer    |
| Performance         | Slower (usually)     | Faster             |
| Best for            | Tree/Divide problems | Iteration problems |

## Deep Concept Understanding

#### Recursion works like:

      Push calls into stack
               ↓
         Reach base case
               ↓
    Return values back step-by-step

# -> Lambda Functions

## 1️⃣ What is a Lambda Function?
### 🔹 Definition

#### A Lambda function is:

#### A small anonymous (nameless) function written in a single line.

#### It is used for short, simple operations.

#### It does NOT use def keyword.

### 2️⃣ Basic Syntax

#### lambda parameters : expression


#### Important rules:

#### • No function name

#### • No return keyword

#### • Only one expression allowed

#### • Automatically returns result

###  Example

```python
square = lambda x: x * x
print(square(5))
```

## 3️⃣ Lambda vs Normal Function
### Normal Function

```python
def add(a, b):
    return a + b

print(add(3, 4))
```

### Lambda Version

```python
add = lambda a, b: a + b
print(add(3, 4))
```

## 4️⃣ Internal Working of Lambda

#### When you write:

```python
lambda x: x * 2
```

#### Python internally treats it like:

```python
def temp(x):
    return x * 2
```

## 5️⃣ Lambda with map()
### 🔹 What is map()?

#### map() applies a function to every element of an iterable.

### Example

```python
numbers = [1, 2, 3, 4]

result = list(map(lambda x: x * 2, numbers))
print(result)
```

## 6️⃣ Lambda with filter()
### 🔹 What is filter()?

#### filter() selects elements based on condition.

### Example

```python
numbers = [1, 2, 3, 4, 5, 6]

even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)
```

## 7️⃣ Lambda with sorted()

#### Very powerful usage.

### Example 1 – Sorting by Second Element

```python
students = [("Ali", 80), ("Sara", 95), ("Ahmed", 70)]

sorted_list = sorted(students, key=lambda x: x[1])
print(sorted_list)
```

### Example 2 – Sort by Name Length

```python
names = ["Ali", "Muhammad", "Sara"]

sorted_names = sorted(names, key=lambda x: len(x))
print(sorted_names)
```

## 8️⃣ Lambda as Argument (Higher-Order Function Concept)

#### A higher-order function:

#### A function that takes another function as argument.

### Example

```python
def operate(func, value):
    return func(value)

result = operate(lambda x: x + 10, 5)
print(result)
```

## 6️⃣ Mutable vs Immutable Behavior (Very Important)
### Mutable Data Types:

#### • List

#### • Dictionary

#### • Set

#### Changes inside function affect original object.

### Immutable Data Types:

#### • int

#### • float

#### • string

#### • tuple

#### Changes inside function do NOT affect original.

### Example

```python
def change_number(n):
    n = 50

x = 10
change_number(x)
print(x)
```

## 7️⃣ Returning Data Structures

#### Functions often return complex structures.

### 🔹 Returning List

```python
def even_numbers(n):
    return [i for i in range(n) if i % 2 == 0]

print(even_numbers(10))
```

### 🔹 Returning Dictionary

```python
def count_letters(word):
    result = {}
    for letter in word:
        result[letter] = result.get(letter, 0) + 1
    return result

print(count_letters("hello"))
```

# -> Advanced Functional Programming Concepts

## 1️⃣ Nested Functions
### 🔹 Definition

#### A nested function is a function defined inside another function.

### Example

```python
def outer():
    print("Outer function")

    def inner():
        print("Inner function")

    inner()

outer()
```

## 2️⃣ Why Use Nested Functions?

#### • Code organization

#### • Encapsulation

#### • Hiding helper logic

#### • Preventing external access

## 3️⃣ Enclosing Scope Recap

#### Nested functions can access variables of outer function.

### Example:

```python
def outer():
    message = "Hello"

    def inner():
        print(message)

    inner()

outer()
```

## 4️⃣ What is a Closure? (Very Important)
### 🔹 Definition

#### A closure is:

#### A function that remembers the variables from its enclosing scope even after the outer function has finished executing.

#### This is a powerful concept.

## 5️⃣ Closure Example

```python
def outer():
    message = "Hello"

    def inner():
        print(message)

    return inner

my_function = outer()
my_function()
```

## 6️⃣ Internal Working of Closure

#### When outer() returns inner:

#### Python creates a closure object that stores:

#### inner function
####     +
#### reference to message variable


#### Even though outer() execution ended,
#### its variables remain stored in closure memory.

## 7️⃣ Closure with Parameters

```python
def multiplier(n):
    def inner(x):
        return x * n
    return inner

double = multiplier(2)
triple = multiplier(3)

print(double(5))
print(triple(5))
```

## 8️⃣ Higher-Order Functions
### 🔹 Definition

#### A higher-order function is a function that:

#### Takes another function as argument
#### OR

#### Returns another function

## 9️⃣ Function as Argument Example

```python
def apply_function(func, value):
    return func(value)

def square(x):
    return x * x

print(apply_function(square, 5))
```

## 🔟 Returning Function Example

```python
def greet():
    def message():
        return "Hello World"
    return message

func = greet()
print(func())
```

## 1️⃣1️⃣ Real-World Practical Example – Access Control

```python
def login_required(func):
    def wrapper():
        print("Checking login...")
        func()
    return wrapper

def dashboard():
    print("Welcome to dashboard")

secure_dashboard = login_required(dashboard)
secure_dashboard()
```

# -> Real-World Case Studies

 ## Case Study 1: Banking System
### 🎯 Problem

#### Design a simple banking system that:

#### • Creates account

#### • Deposits money

#### • Withdraws money

#### • Checks balance

###  • Functional Design

#### We divide system into small functions:

#### create_account()
#### deposit()
#### withdraw()
#### check_balance()


#### Each function has single responsibility.

### Implementation

```python
def create_account(name, balance=0):
    return {"name": name, "balance": balance}

def deposit(account, amount):
    account["balance"] += amount
    return account["balance"]

def withdraw(account, amount):
    if amount > account["balance"]:
        return "Insufficient Funds"
    account["balance"] -= amount
    return account["balance"]

def check_balance(account):
    return account["balance"]

# Usage
account = create_account("Ali", 1000)

deposit(account, 500)
withdraw(account, 300)

print("Final Balance:", check_balance(account))
```

## Case Study 2: E-Commerce Cart System
### 🎯 Problem

#### Build a cart system that:

#### • Adds products

#### • Removes products

#### • Calculates total

#### • Applies discount

### Functional Flow
#### add_to_cart()
#### remove_from_cart()
#### calculate_total()
#### apply_discount()

### Implementation

```python
def add_to_cart(cart, item, price):
    cart[item] = price

def remove_from_cart(cart, item):
    if item in cart:
        del cart[item]

def calculate_total(cart):
    return sum(cart.values())

def apply_discount(total, discount_percent):
    return total - (total * discount_percent / 100)

# Usage
cart = {}

add_to_cart(cart, "Laptop", 1000)
add_to_cart(cart, "Mouse", 50)

total = calculate_total(cart)
final_price = apply_discount(total, 10)

print("Final Price:", final_price)
```

## Case Study 3: Hospital Patient Management System
### 🎯 Problem

#### System should:

#### • Register patient

#### • Add medical record

#### • Display patient info

#### • Calculate bill

###  Functional Flow
#### register_patient()
#### add_record()
#### calculate_bill()
#### display_patient()

### Implementation

```python
def register_patient(name, age):
    return {
        "name": name,
        "age": age,
        "records": [],
        "bill": 0
    }

def add_record(patient, treatment, cost):
    patient["records"].append(treatment)
    patient["bill"] += cost

def calculate_bill(patient):
    return patient["bill"]

def display_patient(patient):
    print("Name:", patient["name"])
    print("Age:", patient["age"])
    print("Treatments:", patient["records"])
    print("Total Bill:", patient["bill"])

# Usage
patient = register_patient("Sara", 30)

add_record(patient, "X-Ray", 200)
add_record(patient, "Blood Test", 100)

display_patient(patient)
```
