# ★ Static Variable :

## ★ Definition :

#### ➝ A static variable is a type of variable that remembers its value even after the function is called multiple time.
#### ➝ It does not reset every time the function runs.
#### ➝ In python, we create it as a function attribute.

# ★ Why Static Variables Are Used

#### ➝ To share data across all objects of a class

#### ➝To retain value between multiple function calls

#### ➝To save memory by using a single copy for all instances

#### ➝To store class-level constants or utility data

# ★ Syntax :

```python
def function_name():
    if not hasattr(function_name, "static_variable"):
        function_name.static_variable = initial_value  # define once
    # modify or use static variable
    function_name.static_variable += 1
```

# ★ For Example  :

```python
def counter():
    if not hasattr(counter, "count"):
        counter.count = 0  # initialize static variable
    counter.count += 1
    return counter.count

print(counter())  # Output: 1
print(counter())  # Output: 2
print(counter())  # Output: 3
```

## ★ Hint to Remember Static Variables :

#### “Static = Same + Stay”

# Tuple :

#### → A tuple is like a list, but it cannot be changed (immutable).

#### → Tuples are ordered, so each item has a position (index).

#### → Tuples can store any data type: numbers, strings, Booleans, or even other tuples.

# 2️⃣ How to Create a Tuple :

```python
# Empty tuple
empty_tuple = ()

# Tuple of numbers
numbers = (1, 2, 3, 4, 5)

# Tuple of strings
fruits = ("apple", "banana", "mango")

# Mixed data types
mixed = (1, "hello", True, 3.14)

# Nested tuple
nested = (1, (2, 3), 4)
```

#### Single item tuple must have a comma 

## Example : 

```python
single = (5,)  # this is a tuple
not_tuple = (5)  # this is just a number
```

## 3️⃣ Access Items

#### Use index (starts from 0)

```python
fruits = ("apple", "banana", "mango")

print(fruits[0])   # apple
print(fruits[2])   # mango

# Negative index
print(fruits[-1])  # mango
print(fruits[-2])  # banana
```

## 4️⃣ Tuples Are Immutable

#### You cannot change items in a tuple.

```python
fruits = ("apple", "banana", "mango")
# fruits[1] = "orange"   ❌ This will give error
```

#### But you can create a new tuple if needed:

```python
fruits = fruits + ("orange",)  # adding item creates new tuple
print(fruits)  # ('apple','banana','mango','orange')
```

## 5️⃣ Tuple Operations

```python
numbers = (1, 2, 3)

# Length
print(len(numbers))   # 3

# Concatenate
a = (1, 2)
b = (3, 4)
print(a + b)          # (1,2,3,4)

# Repeat
print(a * 3)          # (1,2,1,2,1,2)
```

## 6️⃣ Slicing Tuples

```python
numbers = (1, 2, 3, 4, 5)

print(numbers[1:4])  
print(numbers[:3])   
print(numbers[2:])   
print(numbers[-3:])  
```

## 7️⃣ Loop Through Tuple

```python
fruits = ("apple", "banana", "mango")

for fruit in fruits:
    print(fruit)
```

# Type Casting :

#### ➜ Type casting means converting one data type into another.
#### ➜ Simple words: changing the type of data you are working with.

## ★ Syntax

```python
new_variable = datatype(old_variable)
```

## ➜ Example:

```python
x = "10"        # string
y = int(x)      # convert to integer
print(y)        # 10
print(type(y))  # <class 'int'>
```

#### ➜ Python automatically converts the type.

```python
a = 5      # int
b = 2.5    # float
c = a + b
print(c)          # 7.5
print(type(c))    # <class 'float'>
```

#### ★ Key Point: Python converts int to float automatically for accuracy.

#### ➜ Programmer manually converts the type.

```python
x = "25"
y = int(x)
print(y)         
print(type(y))  
```

#### int() → Convert to integer

#### float () → Convert to decimal number

#### str() → Convert to string

#### bool() → Convert to True/False

## ★ Type Casting Important Rules :

#### → Float → Int

#### → Int → Float

#### → Bool Conversion Rules

#### → Input() Always String

#### → Invalid Conversions Cause Error

# ★ Comments in Python

#### ➜ Comments are text written in code to explain it.
#### ➜ Python ignores comments, meaning they are not executed.

# ★ Why Comments Are Important?

#### ➜ Make code readable and understandable.
#### ➜ Help yourself or other programmers to understand the code.
#### ➜ Temporarily disable code for debugging.

## Example:

#### ➜ Comments make your code easier to read and maintain.

#### ➜ Useful for explaining logic or debugging purposes

# ★ Types of Comments

#### 1. Single-line Comments – Use # for a single line explanation

#### 2. Multi-line Comments / Docstrings – Use ''' ''' or """ """ for multiple lines explanation

## ★ 1. Single-line Comments

#### ➜ Start a comment with # for a single line.

```python
# This is a single-line comment
print("Hello World")  # This prints Hello World
```

## ★ 2. Multi-line Comments / Docstrings

#### ➜ Use triple quotes (''' ''' or """ """) for multiple lines.

```python
"""
This is a multi-line comment.
You can write multiple lines of explanation here.
"""
print("Hello")
```

#### ➜ Note: Triple quotes are technically strings, but if not assigned, Python ignores them.

## ★ Important Points

#### ★ Comments are ignored by Python during execution.
#### ★ Use # for single-line comments.
#### ★ Use ''' ''' or """ """ for multi-line comments.
#### ★ Helps to document your code and make it readable.
